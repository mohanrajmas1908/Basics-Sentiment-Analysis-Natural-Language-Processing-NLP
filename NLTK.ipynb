```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk import word_tokenize, sent_tokenize, pos_tag, FreqDist

# Download resources (only once at the top)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')  # WordNet synonyms

# Sample text
text = "I absolutely loved the movie! It was fantastic, but some parts were boring."

# 1. Sentence Tokenization
sentences = sent_tokenize(text)
print("Sentences:", sentences)

# 2. Word Tokenization
words = word_tokenize(text)
print("Words:", words)

# 3. Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words and w.isalpha()]
print("Filtered Words:", filtered_words)

# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(w.lower()) for w in filtered_words]
print("Lemmatized:", lemmatized)

# 5. Stemming
stemmer = PorterStemmer()
stemmed = [stemmer.stem(w.lower()) for w in filtered_words]
print("Stemmed:", stemmed)

# 6. POS Tagging
nltk.download('averaged_perceptron_tagger_eng')
print("POS Tags:", pos_tag)

# 7. Frequency Distribution
freq_dist = FreqDist(lemmatized)
print("Word Frequencies:", freq_dist.most_common())
```
